```python
import os
import urllib
import pickle
import numpy as np
import pandas as pd
import geopandas as gpd

#Plotting
import matplotlib.pyplot as plt
from matplotlib.pyplot import *
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, HoverTool, LogColorMapper
from bokeh.palettes import YlGn as palette #YlGn RdYlGn
from bokeh.models.widgets import Slider, Button, Dropdown, CheckboxGroup, TableColumn, Panel, DataTable, DateFormatter, TableColumn, MultiSelect, Select
from bokeh.io import output_file, show
from bokeh.layouts import Column, widgetbox, Row, Plot
from bokeh.models import ColumnDataSource, GMapOptions, GMapPlot, Patches, HoverTool, CustomJS, DataRange1d
from bokeh.plotting import gmap
palette = palette[9]
palette.reverse()
color_mapper = LogColorMapper(palette=palette)
%matplotlib inline
```

```python
#Download Data
dir_name = "pickles/"

pickles = "https://storage.googleapis.com/w210data/data/pickles/"

if not os.path.isdir(dir_name):
    os.mkdir(dir_name)
    
def download_file(file_name, gcs_bucket):
    full_path = os.path.join(dir_name, file_name)
    if not os.path.exists(full_path):
        urllib.request.urlretrieve (gcs_bucket+file_name, full_path)

def load_file(file_name, gcs_bucket):
    download_file(file_name, gcs_bucket)
    full_path = os.path.join(dir_name, file_name)
    return pd.read_pickle(full_path)

grid = "grid_df.pkl"

grid_df = load_file(grid, pickles)
```

```python
df = pickle.load(open("pickles/grid_df.pkl", "rb"))
df.drop(["halibut_trawl_sites", "marinas", "oil_platforms", "piers"], axis = 1, inplace = True)
df.head()
```

```python
xs = [df["geometry"][i].exterior.xy[0].tolist() for i in range(df.shape[0])]
ys = [df["geometry"][i].exterior.xy[1].tolist() for i in range(df.shape[0])]
biomass_kg = df.biomass.tolist()
water_depth_m = df.depth.tolist()
sst = df.mean_sst.tolist()
species = [df.critical_species[i][0][1] for i in range(df.shape[0])]
critical_species_distance = [df.critical_species[i][0][2] for i in range(df.shape[0])]
county = [df.county[i][0][1] for i in range(df.shape[0])]
county_distance = [df.county[i][0][2] for i in range(df.shape[0])]
pretected_areas = [df.pretected_areas[i][0][1] for i in range(df.shape[0])]
pretected_areas_distance = [df.pretected_areas[i][0][2] for i in range(df.shape[0])]
estab = [df.nes_estab_pct[i][0][1] for i in range(df.shape[0])]
estab_perc = [df.nes_estab_pct[i][0][2] for i in range(df.shape[0])]
avg_emp = [df.annual_avg_emplvl[i][0][1] for i in range(df.shape[0])]
avg_emp_perc = [df.annual_avg_emplvl[i][0][2] for i in range(df.shape[0])]
qcew = [df.qcew_emp_pct[i][0][1] for i in range(df.shape[0])]
qcew_perc = [df.qcew_emp_pct[i][0][2] for i in range(df.shape[0])]
unemployment = [df.unemployment_rate[i][0][1] for i in range(df.shape[0])]
ndvi = df.ndvi.tolist()
min_light = df.z_min_light.tolist()
mixed_l = df.z_mixedl.tolist()
floor_temp = df.floor_temp.tolist()
viable = df.viable.tolist()
boat_launches = [df.boat_launches[i][0][1] for i in range(df.shape[0])]
boat_launches_distance = [df.boat_launches[i][0][2] for i in range(df.shape[0])]
shoretype = [df.shoretype[i][0][1] for i in range(df.shape[0])]
shoretype_distance = [df.shoretype[i][0][2] for i in range(df.shape[0])]
shoretype2 = [df.shoretype2[i][0][1] for i in range(df.shape[0])]
shoretype2_distance = [df.shoretype2[i][0][2] for i in range(df.shape[0])]
aerial_kelp = [df.aerial_kelp[i][0][2] for i in range(df.shape[0])]
admin_kelp = df.admin_kelp_bed.tolist()
```

```python

output_file("combinedplot.html")

map_options = GMapOptions(lat=32.88, lng=-118.460745, map_type="roadmap", zoom=8)

#NEVER SHARE THIS GOOGLE MAPS API KEY
p = gmap("AIzaSyAqMlpFrenE6SClJJ15goq0GmY-MDeN8tg", map_options, title="CA Kelp Areas - Big Squares, Combined Data", height = 600, width = 600)

source = ColumnDataSource(data=dict(
    x = xs,
    y = ys,
    Kelp = biomass_kg,
    Bathymetry = water_depth_m,
    Sea_Surface_Temperature = sst,
    Critical_Species = species,
    Critical_Species_Distance = critical_species_distance,
    County = county,
    County_Distance = county_distance,
    Protected_Areas = pretected_areas,
    Protected_Areas_Distance = pretected_areas_distance,
    Establistment_Percentage = estab,
    Average_Employment_Rate = avg_emp,
    Qcew = qcew,
    Unemployment = unemployment,
    NDVI = ndvi,
    Min_light = min_light,
    Mixed_Layer = mixed_l,
    Floor_Temperature = floor_temp,
    Lease_Viablility = viable,
    Boat_Launches = boat_launches,
    Boat_Launches_Distance = boat_launches_distance,
    Shoretype = shoretype,
    Shoretype_Distance = shoretype_distance,
    Shoretype2 = shoretype2,
    Shoretype2_Distance = shoretype2_distance,
    Aerial_Kelp_Survey = aerial_kelp,
    Admin_Kelp = admin_kelp))


patched = Patches(xs="x", ys="y", fill_color= {"field": "Kelp", "transform":color_mapper}, 
          fill_alpha=0.8, line_color = "white", line_width = 0.7)
p.add_glyph(source, patched)
tooltip_values = ['Kelp','Bathymetry','Sea_Surface_Temperature','Critical_Species', 'Critical_Species_Distance',
                  'County', 'County_Distance', 'Protected_Areas', 'Protected_Areas_Distance','Establistment_Percentage',
                  'Average_Employment_Rate', 'Qcew', 'Unemployment', 'NDVI', 'Min_light', 'Mixed_Layer', 
                  'Floor_Temperature', 'Lease_Viablility', 'Boat_Launches', 'Boat_Launches_Distance', 'Shoretype',
                  'Shoretype_Distance', 'Shoretype2', 'Shoretype2_Distance', 'Aerial_Kelp_Survey','Admin_Kelp']


hover = HoverTool(tooltips=[])
p.add_tools(hover)
########################################################################
multi_select = MultiSelect(title="Metrics Hover:",
                           value=["@Kelp"],
                           options=[(tooltip_values[i], tooltip_values[i].replace("_", " ")) for i in range(len(tooltip_values))])

callback_m = CustomJS(args=dict(hover=hover), code="""
    hover.tooltips = []
    const value = cb_obj.value
    for (i=0; i<value.length; ++i){
        const name = value[i]
        hover.tooltips.push([name.replace(/_/g,' '), "@"+name])
    }
    """)

multi_select.js_on_change('value', callback_m)
########################################################################
dropdown_metrics = ['Kelp','Bathymetry','Sea_Surface_Temperature', 'NDVI', 'Min_light', 'Mixed_Layer', 
                  'Floor_Temperature', 'none', 'Critical_Species_Distance', 'County_Distance', 
                  'Protected_Areas_Distance','Establistment_Percentage', 'Average_Employment_Rate', 
                  'Qcew', 'Unemployment','Lease_Viablility', 'Boat_Launches', 'Boat_Launches_Distance',
                  'Shoretype_Distance', 'Shoretype2_Distance', 'Aerial_Kelp_Survey','Admin_Kelp']

menu = [(i.replace("_", " "), i) for i in dropdown_metrics]
menu[np.where(np.asarray(menu) == 'none')[0][0]] = None
dropdown = Dropdown(label="Metric Color Selection", value = "Kelp", button_type="warning", menu=menu)

callback_d = CustomJS(args=dict(patched=patched, source=source), code="""
    patched.fill_color.field = cb_obj.value;    
    patched.change.emit();
    """)

dropdown.js_on_change("value", callback_d)

show(Row(Column(multi_select, dropdown), p))
```

```python
#Side sliders for any of the metrics
from bokeh.models.glyphs import VBar
from bokeh.plotting import figure, show, output_file

n = figure(title="NDVI Distribution", height = 200, width = 200)
hist, edges = np.histogram(ndvi, density=True, bins=5)
source = ColumnDataSource(data=dict(edges = edges, hist = hist, all_edges = edges))

glyph = VBar(x="edges", top="hist", bottom = 0, width = 0.0022121102435833442)
n.add_glyph(source, glyph)
n.xaxis.axis_label = "NDVI"
n.yaxis.axis_label = "Frequency"


slider = Slider(start=min(edges), end=max(edges), value=0, step=.01, title="NDVI Minimum")

update_curve = CustomJS(args=dict(source=source, slider=slider), code="""
    var data = source.data;
    console.log("slider value", slider.value)
    var f = slider.value;
    console.log("prefilter data", data)
    data["edges"] = data["all_edges"].filter(function(x) {
    return x > f;
    });
    console.log("after filter data", data["edges"], data["all_edges"])
    source.change.emit();
    """)
slider.js_on_change('value', update_curve)

show(Column(slider, n))
show(Row(Column(multi_select, dropdown), p))
```
